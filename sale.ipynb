```python
import pandas as pd

file_path = '/content/drive/MyDrive/group/retail_features_processed.csv'
df = pd.read_csv(file_path)
df.head()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, Dense, Dropout, BatchNormalization, Layer
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import tensorflow.keras.backend as K

# 自定义 Attention 层
class Attention(Layer):
    def __init__(self):
        super(Attention, self).__init__()

    def call(self, inputs):  # inputs shape: (batch, time_steps, features)
        score = K.softmax(K.sum(inputs, axis=-1, keepdims=True), axis=1)  # 注意力权重
        context = inputs * score
        context_vector = K.sum(context, axis=1)  # 加权求和
        return context_vector
# ----------------------------
# 1. 特征定义 + 预处理
# ----------------------------
num_features = ['sell_price', 'lag_1', 'lag_7', 'lag_28',
                'rolling_mean_7', 'rolling_std_7',
                'rolling_mean_28', 'rolling_max_28']
target = 'sales'

df = df.dropna()

# 标准化数值特征
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[num_features])

# 提取目标变量，并做 log1p 变换
y = df[target].values
log1p_sales = np.log1p(y).reshape(-1, 1)

# 归一化目标变量
target_scaler = MinMaxScaler()
y_scaled = target_scaler.fit_transform(log1p_sales).flatten()


# 特征和目标定义
num_features = ['sell_price', 'lag_1', 'lag_7', 'lag_28',
                'rolling_mean_7', 'rolling_std_7',
                'rolling_mean_28', 'rolling_max_28']
target = 'sales'

df = df.dropna()

# 特征归一化（StandardScaler 更适合 LSTM）
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df[num_features])

# 目标变量 log1p + MinMaxScaler
y = df[target].values
log1p_sales = np.log1p(y).reshape(-1, 1)
target_scaler = MinMaxScaler()
y_scaled = target_scaler.fit_transform(log1p_sales).flatten()

# 构建序列
def create_sequences(X, y, seq_length=28):
    X_seq, y_seq = [], []
    for i in range(len(X) - seq_length):
        X_seq.append(X[i:i+seq_length])
        y_seq.append(y[i+seq_length])
    return np.array(X_seq), np.array(y_seq)

X_seq, y_seq = create_sequences(X_scaled, y_scaled)
X_train, X_test, y_train, y_test = train_test_split(X_seq, y_seq, test_size=0.2, random_state=42)

# ----------------------------
# 3. 构建 LSTM 模型
# ----------------------------
# 构建模型结构
input_layer = Input(shape=(X_train.shape[1], X_train.shape[2]))

x = LSTM(128, return_sequences=True)(input_layer)
x = BatchNormalization()(x)
x = Dropout(0.2)(x)

x = LSTM(64, return_sequences=True)(x)
x = BatchNormalization()(x)
x = Dropout(0.2)(x)

x = LSTM(32, return_sequences=True)(x)

# Attention 层聚焦重要时间步
x = Attention()(x)
x = Dense(64, activation='relu')(x)
output = Dense(1, activation='sigmoid')(x)  # sigmoid 匹配 MinMaxScaler 范围

model = Model(inputs=input_layer, outputs=output)
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()


# ----------------------------
# 4. 模型训练
# ----------------------------
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
lr_reduce = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, min_lr=1e-5)

history = model.fit(
    X_train, y_train,
    epochs=50,
    batch_size=32,
    validation_split=0.2,
    callbacks=[early_stop, lr_reduce],
    verbose=1
)


# 模型评估
loss, mae = model.evaluate(X_test, y_test)
print(f"Test MAE (scaled): {mae:.4f}")

# 反归一化预测结果
y_pred_scaled = model.predict(X_test)
y_pred = target_scaler.inverse_transform(y_pred_scaled)
y_true = target_scaler.inverse_transform(y_test.reshape(-1, 1))

# 可视化 Loss 曲线
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss Curve')
plt.legend()
plt.grid(True)
plt.show()

# 可视化 真实 vs 预测
plt.figure(figsize=(12, 6))
plt.plot(y_true[:200], label='True')
plt.plot(y_pred[:200], label='Predicted')
plt.title('Predicted vs True Sales')
plt.xlabel('Sample')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```
